```python
logging.basicConfig(filename = 'test2.log', level = logging.DEBUG, format = '%(asctime)s %(name)s %(levelname)s %(message)s')
```

```python
l = [1,2,3,4,[2,3,4],'nabin','katwal']
```

```python
l1_int = []
l2_str = []
for i in l:
    logging.info('we are iterating through our list and our local variable is {}'.format(l))
    if type(i) == list:
        logging.info('i am inside if statement and i am trying to check list type'+str(i))
        for j in i:
            logging.info('i am in another for loop for list inside list element')
            if type(j) == int:
                logging.info('i am inside if statement')
                l1_int.append(j)
    elif type(i) == int:
        l1_int.append(i)
    else:
        if type(i) == str:
            l2_str.append(i)
            logging.info('my final result for int is {l1} and str is {l2}'.format(l1 = l1_int, l2 = l2_str))
```
