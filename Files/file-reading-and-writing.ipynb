```python
import csv
```

```python
data = [['name','email_id','phone_number'],
        ['nabin','nk@gmail.com',9853321],
        ['milan','m@gmail.com',9822114]
       ]
```

```python
with open('data.csv','w') as f:
    writer = csv.writer(f)
    
    for i in data:
        writer.writerow(i)
```

```python
with open('data.csv','r') as f:
    read_data = csv.reader(f)
    
    for i in read_data:
        print(i)
```
