```python
import time
```

```python
def test1(id):
    for i in range(10):
        print('test1 %d printing %d' %(id,i))
        time.sleep(1)
```

```python
test1(1)
```

```python
test1(0)
```

```python
thread1 = [threading.Thread(target = test1, args = (i)) for i in range(3)]
```

```python
for t in thread1:
    t.start()
```

```python
def test2(id):
    for i in range(10):
        print('test2 %d printing %d %s' %(id,i,time.ctime()))
        time.sleep(1)
```

```python
test2(0)
```

```python
import threading
shared_variable = 0
lock_variable = threading.Lock()
```

```python
def test3(id):
    global shared_var
    with lock_var:
        shared_var = shared_var + 1
        print('test3 id is %d has increased the shared variable by %d' %(id,shared_var))
```

```python
 ther3 = [threading.Thread(target = test2, args = (id) for i in range(3))]
```

```python
ther3
```

```python
for t in ther3:
    t.start()
```
