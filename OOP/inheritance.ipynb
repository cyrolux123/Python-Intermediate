```python
class test:
    def test_meth(self):
        return 'this is my first class'
```

```python
class child_test(test):
    pass
```

```python
child_test_obj = child_test()
```

```python
child_test_obj.test_meth()
```

```python
class class1:
    def test_class1(self):
        return 'this is a method from class1'
```

```python
class class2(class1):
    def test_class2(self):
        return 'this is a method from class2'
```

```python
class class3(class2):
    pass
```

```python
obj_class3 = class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```

```python
class father:
    def father_property(self):
        return 'this is the property of father'
```

```python
class mother:
    def mother_property(self):
        return 'this is the property of mother'
```

```python
class child(father,mother):
    pass
```

```python
obj_child = child()
```

```python
obj_child.father_property()
```

```python
obj_child.mother_property()
```
