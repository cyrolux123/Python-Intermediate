```python
class csit2:
    def student_details(self, name, mail_id, number):
        print(name,mail_id,number)
        
    @staticmethod    
    def mentor_id(mail_id_mentor):
        print(mail_id_mentor)
        
    @staticmethod
    def mentor(list_mentor):
        csit2.mentor_id(['n@gmail.com','k@gmail.com'])
        print(list_mentor)
        
    @classmethod
    def class_name(cls):
        cls.mentor(['nabin','katwal'])
        
    def mentorr(self,mentor_list):
        print(mentor_list)
        self.mentor(['milan','arjun'])
```
