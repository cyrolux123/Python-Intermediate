```python
class csit:
    def __init__(self,name,email):
        self.name = name
        self.email = email
        
    def student_details(self):
        print(self.name, self.email)
```

```python
class bsccsit:
    def __init__(self,name,email):
        self.name = name
        self.email = email
        
    @classmethod    
    def details(cls, name, email):
        return cls(name, email)
    
    def student_details(self):
        print(self.name, self.email)
```

```python
bsc = bsccsit.details('miraj','miraj@gmail.com')
```

```python
bsc.name
```

```python
bsc.email
```

```python
bsc.student_details()
```
