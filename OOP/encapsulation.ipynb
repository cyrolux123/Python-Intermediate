**Encapsulation helps to keep entities private so other cannot see internal structure and modify**

```python
class car: #__ makes variable private which can't be directly accessed
    def __init__(self,year,make,model,speed): 
        self.__year = year
        self.__make = make
        self.__model = model
        self.__speed = 0
        
    def set_speed(self,speed):
        self.__speed = 0 if speed < 0 else speed
        
    def get_speed(self):
        return self.__speed
```

```python
c = car(2024,'toyota','innova','12')
```

```python
c._car__year
```

```python
c.set_speed(-4444)
```

```python
c.get_speed()
```

```python
c.set_speed(1234)
```

```python
c.get_speed()
```

```python
class bank_account:
    def __init__(self,balance):
        self.__balance = balance
        
    def deposit(self, amount):
        self.__balance = self.__balance + amount
        
    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance = self.__balance - amount
            return True
        else:
            return False
    
    def get_balance(self):
        return self.__balance
```

```python
diwas = bank_account(1000)
```

```python
diwas.get_balance()
```

```python
diwas.deposit(5000)
```

```python
diwas.get_balance()
```

```python
diwas.deposit(1000)
```

```python
diwas.get_balance()
```

```python
diwas.withdraw(9000)
```

```python
diwas.withdraw(2000)
```
