```python
import abc
class csit:
    
    @abc.abstractmethod
    def student_details(self):
        pass
    
    @abc.abstractmethod
    def student_assignment(self):
        pass
    
    @abc.abstractmethod
    def student_marks(self):
        pass
```

```python
class student_details(csit):
    def student_details(self):
        return 'this is a method for taking student details'
    
    def student_assignment(self):
        return 'this is a method for assignment details for a particular student'
```

```python
class first_sem(csit):
    def student_details(self):
        return 'this will return the details for first sem'
    def student_assignment(self):
        return 'this will give you assignment details for first sem'
```

```python
fs = first_sem()
fs.student_details()
```
