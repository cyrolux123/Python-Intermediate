```python
def test():
    print('this is the start of my function')
    print('this is my function to test')
    print(4+5)
    print('this is the end of my function')
```

```python
test()
```

```python
def decorate(func):
    def inner_decorate():
        print('this is the start of my function')
        func()
        print('this is the end of my function')
    return inner_decorate
```

```python
@decorate
def test1():
    print(6+7)
```

```python
test1()
```

```python
import time

def timer_test(func):
    def timer_test_inner():
        start = time.time()
        func()
        end = time.time()
        print(end - start)
    return timer_test_inner
```

```python
def test2():
    print(45+78)
```

```python
test2()
```

```python
@timer_test
def test3():
    print(45+78)
```

```python
test3()
```

```python
@timer_test
def test4():
    for i in range(1000):
        pass
```

```python
test4()
```
